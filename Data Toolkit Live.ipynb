1. **Demonstrate three different methods for creating identical 2D arrays in NumPy Provide the code for each method and the final output after each method**

```python
# Method 1: Using np.full()
import numpy as np

# Create a 2D array of size 3x4 filled with 5s
array_1 = np.full((3, 4), 5)

# Print the array
print("Array created using np.full():")
print(array_1)

# Method 2: Using np.ones() and np.multiply()
array_2 = np.ones((3, 4)) * 5

# Print the array
print("\nArray created using np.ones() and np.multiply():")
print(array_2)

# Method 3: Using np.tile()
array_3 = np.tile([5], (3, 4))

# Print the array
print("\nArray created using np.tile():")
print(array_3)
```

3. **Explain the following terms:**  
**• The difference in np.array, np.asarray and np.asanyarray.**   
**• The difference between Deep copy and shallow copy.**


**Difference between np.array, np.asarray and np.asanyarray:**

- `np.array`:
    - Creates a new array object from the given input.
    - Can take a variety of input types, including lists, tuples, and other arrays.
    - Always creates a copy of the input data.
- `np.asarray`:
    - Converts the input data to an array object if it is not already an array.
    - If the input data is already an array, it will not create a new copy of the data.
    - Can take a variety of input types, including lists, tuples, and other arrays.
- `np.asanyarray`:
    - Similar to `np.asarray`, but it will try to convert the input data to an array object even if it is not a sequence type.
    - For example, it can convert a scalar value to a 0-dimensional array.
    - Can take a variety of input types, including lists, tuples, scalars, and other arrays.

**Difference between Deep copy and shallow copy:**

- **Shallow copy**:
    - Creates a new object that contains references to the original object's data.
    - Changes made to the new object will also affect the original object.
    - Can be performed using the `copy()` method or the `[:]` slice operator.
- **Deep copy**:
    - Creates a new object that contains a completely independent copy of the original object's data.
    - Changes made to the new object will not affect the original object.
    - Can be performed using the `deepcopy()` method.

6.**Create a 3D NumPy array of shape (3, 3, 3) containing random integers between 1 and 10. Perform the following operations:**    
a) **Find the indices of the maximum values along each depth level (third axis).**   
b) **Perform element-wise multiplication of between both array**.

```python
array = np.random.randint(1, 11, size=(3, 3, 3))

# Find the indices of the maximum values along each depth level (third axis)
max_indices = np.unravel_index(np.argmax(array, axis=None), array.shape)

# Perform element-wise multiplication of between both array
result = array * array

print("Original array:")
print(array)

print("\nIndices of maximum values:")
print(max_indices)

print("\nElement-wise multiplication result:")
print(result)
```

15. **Create a NumPy array data containing 1000 samples from a normal distribution. Perform the following tasks using Matplotlib:**   
a) **Plot a histogram of the data with 30 bins.**    
b) **Overlay a line plot representing the normal distribution's probability density function (PDF).**    
c) **Label the x-axis as 'Value' and the y-axis as 'Frequency/Probability'.**       
d) **Set the title of the plot as 'Histogram with PDF Overlay'.**

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Create a NumPy array containing 1000 samples from a normal distribution
data = np.random.randn(1000)

# Plot a histogram of the data with 30 bins
plt.hist(data, bins=30, density=True, alpha=0.6, color='g')

# Overlay a line plot representing the normal distribution's PDF
xmin, xmax = plt.xlim()
x = np.linspace(xmin, xmax, 100)
p = norm.pdf(x, 0, 1)
plt.plot(x, p, 'k', linewidth=2)

# Label the x-axis and y-axis
plt.xlabel('Value')
plt.ylabel('Frequency/Probability')

# Set the title of the plot
plt.title('Histogram with PDF Overlay')

# Show the plot
plt.show()
```

**17. Create a Seaborn scatter plot of two random arrays, color points based on their position relative to the origin (quadrants), add a legend, label the axes, and set the title as 'Quadrant-wise Scatter Plot'.**

```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# Create two random arrays
x = np.random.randn(100)
y = np.random.randn(100)

# Create a DataFrame
df = pd.DataFrame({'x': x, 'y': y})

# Define a function to determine the quadrant
def quadrant(x, y):
    if x >= 0 and y >= 0:
        return 'Quadrant I'
    elif x < 0 and y >= 0:
        return 'Quadrant II'
    elif x < 0 and y < 0:
        return 'Quadrant III'
    else:
        return 'Quadrant IV'

# Apply the function to the DataFrame
df['quadrant'] = df.apply(lambda row: quadrant(row['x'], row['y']), axis=1)

# Create a scatter plot
sns.scatterplot(data=df, x='x', y='y', hue='quadrant', palette='bright')

# Label the axes
plt.xlabel('X')
plt.ylabel('Y')

# Set the title
plt.title('Quadrant-wise Scatter Plot')

# Show the plot
plt.show()
```

**18. With Bokeh, plot a line chart of a sine wave function, add grid lines, label the axes, and set the title as 'Sine Wave Function'.**

```python
from bokeh.plotting import figure, show
from bokeh.io import output_notebook
import numpy as np

# Prepare the data
x = np.linspace(0, 4*np.pi, 100)
y = np.sin(x)

# Create a new plot with a title and axis labels
p = figure(title="Sine Wave Function", x_axis_label='x', y_axis_label='sin(x)')

# Add a line renderer with legend and line thickness
p.line(x, y, legend_label="sin(x)", line_width=2)

# Show the results
show(p)
```

**19. Using Bokeh, generate a bar chart of randomly generated categorical data, color bars based on their values, add hover tooltips to display exact values, label the axes, and set the title as 'Random Categorical Bar Chart'.**

```python
from bokeh.models import HoverTool
from bokeh.palettes import Spectral6
from bokeh.plotting import figure, show
from bokeh.transform import factor_cmap
import numpy as np

# Prepare the data
categories = ['A', 'B', 'C', 'D', 'E', 'F']
values = np.random.randint(1, 10, size=len(categories))

# Create a new plot with a title and axis labels
p = figure(x_range=categories, title="Random Categorical Bar Chart", x_axis_label='Category', y_axis_label='Value')

# Add a bar renderer with legend and color mapped to the categories
p.vbar(x=categories, top=values, width=0.9, legend_field="x", line_color='white',
       fill_color=factor_cmap('x', palette=Spectral6, factors=categories))

# Add hover tooltips
hover = HoverTool(tooltips=[("Value", "@top")])
p.add_tools(hover)

# Show the results
show(p)
```
